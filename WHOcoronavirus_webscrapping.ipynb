```python
#importing libraries the needed packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def convert_str_to_num(val):
    if val=='':
        val=0
    else:
        val=float(val) 
    return val   
```

```python
# get content from more than 1 page
content_lst=[]
tag_lst=[]
# for i in range(1,num_pages+1):
url=f"https://covid19.who.int/table" #page
response=requests.get(url)
response=response.content
soup=BeautifulSoup(response,'html.parser')#html parser will change the format of above content to html
ol = soup.find('div',class_='tbody')
contents = ol.find_all('div',class_='tr depth_0')
tag_lst=[]
for content in contents:
#     if cost.attrs['class']==['pad5bottom', 'small'] and cost.text!=" ":
        print(content)
        name=content.find('span').text
        content= content.find_all('div',class_='sc-pXZzD fMTLAh')
        total_cases=content[0].text.replace(',','')
        total_cases=convert_str_to_num(total_cases)
        new_cases=content[1].text.replace(',','')
        new_cases=convert_str_to_num(new_cases)
        total_deaths=content[2].text.replace(',','')
        total_deaths=convert_str_to_num(total_deaths)
        new_death=content[3].text.replace(',','')
        new_death=convert_str_to_num(new_death)
        vaccine_dose=content[4].text.replace(',','') #Total vaccine doses administered per 100 population
        vaccine_dose=convert_str_to_num(vaccine_dose)
        vaccinated=content[5].text.replace(',','') #vaccinated with last dose of primary series per 100 population
        vaccinated=convert_str_to_num(vaccinated)
        boosted=content[6].text.replace(',','') #Persons Boosted per 100 population
        boosted=convert_str_to_num(boosted)
        content_lst.append([name,total_cases,new_cases,total_deaths,new_death,vaccine_dose,vaccinated,boosted])
        print("tc: ",total_cases)
        print("nc: ",new_cases)
        print("td: ",total_deaths)
        print("nc: ",new_death)
        print("vd: ",vaccine_dose)
        print("v: ",vaccinated)
        print("b: ",boosted)
        print()
```

```python
df= pd.DataFrame(content_lst,columns=['Country','Total_cases','New_cases','Total_deaths','New_deaths','Total vaccine doses administered per 100 population','fully vaccinated with last dose of primary series per 100 population','Persons Boosted per 100 population'])
```

```python
# importing modules
import requests
from bs4 import BeautifulSoup

# URL for scraping data
url = 'https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/'

# get URL html
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

data = []

# soup.find_all('td') will scrape every
# element in the url's table
data_iterator = iter(soup.find_all('td'))

# data_iterator is the iterator of the table
# This loop will keep repeating till there is
# data available in the iterator
while True:
	try:
		country = next(data_iterator).text
		confirmed = next(data_iterator).text
		deaths = next(data_iterator).text
		continent = next(data_iterator).text

		# For 'confirmed' and 'deaths',
		# make sure to remove the commas
		# and convert to int
		data.append((
			country,
			int(confirmed.replace(',', '')),
			int(deaths.replace(',', '')),
			continent
		))

	# StopIteration error is raised when
	# there are no more elements left to
	# iterate through
	except StopIteration:
		break

# Sort the data by the number of confirmed cases
data.sort(key = lambda row: row[1], reverse = True)
```

```python
# create texttable object
import texttable as tt
table = tt.Texttable()

# Add an empty row at the beginning for the headers
table.add_rows([(None, None, None, None)] + data)

# 'l' denotes left, 'c' denotes center,
# and 'r' denotes right
table.set_cols_align(('c', 'c', 'c', 'c'))
table.header((' Country ', ' Number of cases ', ' Deaths ', ' Continent '))

print(table.draw())
```
